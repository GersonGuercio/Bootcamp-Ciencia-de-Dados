## Quanto cada parlamenta gasta?

```python
data_2018.groupby(['txNomeParlamentar'])['vlrDocumento'].sum().sort_values(ascending=False)[:10]
```

## Quem gastou mais vezes?

```python
data_2018['txNomeParlamentar'].value_counts()[:10]
```

## Quantos deputados/ agentes públicos tem por partido?

```python
partidos = data_2018.groupby(['sgPartido'])['txNomeParlamentar'].nunique()
```

## Gasto por partido por agentes públicos

```python
gasto = data_2018.groupby(['sgPartido'])['vlrDocumento'].sum()
```

```python
df = pd.concat([gasto, partidos], axis=1)
df["gastosDeputados"] = df["vlrDocumento"] / df["txNomeParlamentar"]
df.sort_values(ascending=False, by='gastosDeputados')[:10]
```
